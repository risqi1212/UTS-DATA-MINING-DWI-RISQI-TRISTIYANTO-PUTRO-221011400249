```python
#1 Please get the list of all Companies and sort by Company Domain in reverse order.

# Fungsi untuk mengurutkan perusahaan berdasarkan domain secara terbalik
def sort_company(company_detail_list):
    # Ekstrak nama dan domain perusahaan ke dalam list dictionary
    companies = [{'name': company['name'], 'domain': company['domain']} 
                 for company in company_detail_list.values()]
    
    # Urutkan perusahaan berdasarkan domain secara terbalik
    sorted_companies = sorted(companies, key=lambda x: x['domain'], reverse=True)
    return sorted_companies

# Data detail perusahaan
company_detail_list = {
    'Company 1': {
        'name': 'Company 1',
        'domain': 'Retail',
        # ... (data lainnya)
    },
    'Company 2': {
        'name': 'Company 2',
        'domain': 'Construction',
        # ... (data lainnya)
    },
    'Company 3': {
        'name': 'Company 3',
        'domain': 'Healthcare',
        # ... (data lainnya)
    }
}

# Contoh penggunaan fungsi:
sorted_companies = sort_company(company_detail_list)
for company in sorted_companies:
    print(company)
```

```python
#3 Create a function that return employees with it's company domain.

def get_employees(employee_list, company_details):
    # Menggunakan list comprehension untuk membuat daftar karyawan dengan domain perusahaan mereka
    return [
        {
            'full_name': employee['full_name'],
            'company': employee['company'],
            'domain': company_details[employee['company']]['domain']
        }
        for employee in employee_list.values()
    ]

# Data karyawan
employee_detail_list = {
    'John Doe': {
        'name': 'EMP-0001',
        'first_name': 'John',
        'last_name': 'Doe',
        'full_name': 'John Doe',
        'company': 'Company 1'
    },
    'Tom Smith': {
        'name': 'EMP-0002',
        'first_name': 'Tom',
        'last_name': 'Smith',
        'full_name': 'Tom Smith',
        'company': 'Company 2'
    },
    'Andrew Sebastian': {
        'name': 'EMP-0003',
        'first_name': 'Andrew',
        'last_name': 'Sebastian',
        'full_name': 'Andrew Sebastian',
        'company': 'Company 2'
    },
}

# Data detail perusahaan
company_detail_list = {
    'Company 1': {
        'name': 'Company 1',
        'domain': 'Retail',
        'clients': [
            {'name': 'acme.inc', 'country': 'united states'},
            {'name': 'Wayne.co', 'country': 'united states'}
        ]
    },
    'Company 2': {
        'name': 'Company 2',
        'domain': 'Construction',
        'clients': [
            {'name': 'Tesla', 'country': 'united states'},
            {'name': 'Japan Airlines', 'country': 'japan'},
            {'name': 'Indofood', 'country': 'indonesia'}
        ]
    },
    'Company 3': {
        'name': 'Company 3',
        'domain': 'Healthcare',
        'clients': [
            {'name': 'Petronas', 'country': 'malaysia'},
            {'name': 'VW Group', 'country': 'germany'},
            {'name': 'IBM', 'country': 'united states'},
            {'name': 'Mitsubishi', 'country': 'japan'}
        ]
    }
}

# Memanggil fungsi untuk mendapatkan daftar karyawan dengan domain perusahaan
employee_domain_list = get_employees(employee_detail_list, company_detail_list)

# Mencetak setiap item dalam daftar pada baris baru
for employee in employee_domain_list:
    print(f"Full Name: {employee['full_name']}")
    print(f"Company: {employee['company']}")
    print(f"Domain: {employee['domain']}\n")
```
